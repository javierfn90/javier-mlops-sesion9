En el código anterior se usa la función "rename" de pandas junto con una función lambda.


1.   str.lower() convierte todos los caracteres a minúsculas.
2.   str.replace(' ', '_') reemplaza todos los espacios en blanco con guiones bajos.
1.   str.replace('\W', '') reemplaza todos los caracteres que no sean palabras (es decir, caracteres que no sean alfanuméricos o guiones bajos) con nada.
